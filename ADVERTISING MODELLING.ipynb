# MODELLING

## Decision trees model

creating a supervised learning model to help identify which individuals are most likely to click on the ads in the blog. 

```r
# Previewing the data.
adv_new
```

```r
# subsetting our data to contain labels and features.
adv_suset = adv_new[, c(1,2,3,4,7,10)]
head(adv_suset)
```

```r
# Decision Trees algorithm will be used to do this classification 
#
# Importing library rpart 
#
library(rpart)
#install.packages('rpart.plot')
library(rpart.plot)

# Training the model
#
model1 <- rpart(Clicked.on.Ad ~ ., data = adv_suset ,
           method = "class")
```

```r
# Plotting the results.
#
rpart.plot(model1)
```

```r
# Prediction.
preds1 <- predict(model1, adv_suset, type = "class")
table(preds1, adv_suset$Clicked.on.Ad)
```

The above confusion matrix shows that the mode performed fairly well in classifying the independent variable

```r
mean(adv_suset$Clicked.on.Ad == preds1)
```

The decision trees algorithm has performed well in classification with an accuracy of 96%
